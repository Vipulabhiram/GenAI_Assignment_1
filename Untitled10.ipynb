```python
#importing pipeline function from transformers Library
from transformers import pipeline
summarization = pipeline('summarization',model="facebook/bart-large-cnn")
zeroShortPrompt = """India is a vast and diverse country located in South Asia. Bordered by Pakistan to the west, China, Nepal, and Bhutan to the north, and Bangladesh and Myanmar to the east, India is surrounded by the Indian Ocean to the south. It is the seventh-largest country in the world by area and the most populous democracy. India has a federal parliamentary democratic government, with New Delhi serving as the national capital. The President is the ceremonial head of state, while the Prime Minister is the head of government. The country is divided into 28 states and 8 union territories, each with its own government.
"""
print(summarization(zeroShortPrompt))
```

```python
transulation = pipeline("translation",model="Helsinki-NLP/opus-mt-en-hi")
fewShortPrompt = """Hi My  name is Vipul:हाय मेरा नाम वुल्फ्व है
What is your name:  """
print(transulation(fewShortPrompt))
```
