```python
HF_TOKEN = getpass("Enter your HF TOKEN:")
API_URL = "https://api-inference.huggingface.co/models/deepset/roberta-base-squad2"
headers = {
    "Authorization": f"Bearer {HF_TOKEN}"
}
```

```python
def question_answering(question, context):
    payload = {
        "inputs": {
            "question": question,
            "context": context
        }
    }
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()
```

```python
question = "When is independence day?"
context = """
Independence Day is a national festival celebrated with great pride and enthusiasm across India on August 15th every year. It marks the historic day in 1947 when India gained freedom from British colonial rule after decades of struggle and sacrifice by countless freedom fighters. On this day, the Prime Minister hoists the national flag at the Red Fort in Delhi and addresses the nation, honoring the spirit of patriotism and unity. Schools, colleges, and public institutions organize cultural events, parades, and flag-hoisting ceremonies to pay tribute to the nation's journey toward independence. It is a day to remember our past, celebrate our present, and commit ourselves to building a better future for the country.

"""
```

```python

output = question_answering(question, context)
print("Answer:", output.get("answer"))
```

```python
# Import necessary libraries
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
def generate_story(prompt, temperature=0.7):
    try:
        # Load pre-trained model and tokenizer
        tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
        model = GPT2LMHeadModel.from_pretrained("gpt2")

        # Encode the input prompt
        input_ids = tokenizer.encode(prompt, return_tensors="pt")

        # Generate output using temperature and sampling parameters
        output = model.generate(
            input_ids,
            do_sample=True,
            temperature=temperature,
            top_k=50,
            top_p=0.95,
            max_length=150,
            repetition_penalty=1.2,
            pad_token_id=tokenizer.eos_token_id  # ✅ This prevents warnings
        )

        # Decode and print the generated text
        story = tokenizer.decode(output[0], skip_special_tokens=True)
        print(f"\n--- Temperature: {temperature} ---\n{story}\n")

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
prompt = "Once upon a time "
temperatures = [0.2, 0.7, 1.1, 1.3]

for temp in temperatures:
    generate_story(prompt, temperature=temp)
```
